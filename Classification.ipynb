```python
def avg(wv,word):
    features = np.zeros(300,dtype = 'float64')
    word = word.split(' ')
    count = 0
    for i in word:
        try:
            count+=1
            features = features+wv[i]
        except:
            pass
    features = np.array(features)
    print(count)
    features = np.divide(features,count)
    return features
```
